```python
df=pd.read_csv('c:\\Users\\User\\Desktop\\Financial Data specialist\\GIT AND GITHUB\\AGRI-CREDIT ASSESMENT MODEL\\DATA\\DATA-EXPLORATION\\EA_Agriculture_data_final.csv')
```

```python
df=df.drop(columns='Unnamed: 0')
```

```python
df.tail()
```

```python
X=df.drop(columns='default_probability')
y=df['default_probability']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_other,X_test,y_other,y_test=train_test_split(X,y,test_size=0.1,random_state=101)
X_train,X_eval,y_train,y_eval=train_test_split(X_other,y_other,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_eval=scaler.transform(X_eval)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import RidgeCV
```

```python
#alphas to try
alphas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
```

```python
#RidgeCV model
ridge_cv = RidgeCV(alphas=alphas, cv=5)
```

```python
ridge_cv.fit(X_train,y_train)
```

```python
# Check which alpha was chosen
print(f"Best alpha: {ridge_cv.alpha_}")
```

```python
# Predict on validation set
y_val_pred = ridge_cv.predict(X_eval)
```

```python
#Evaluation
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_eval,y_val_pred)
rmse=np.sqrt(mse)
```

```python
mean_pred=df['default_probability'].mean()
rmse/mean_pred*100
```

```python
final_y_pred=ridge_cv.predict(X_test)
```

```python
final_mse=mean_squared_error(y_test,final_y_pred)
final_rmse=np.sqrt(final_mse)
```
